The Rossmann Stores Data is a comprehensive dataset provided by Rossmann, a large chain of drug stores in Europe. This dataset is primarily used to analyze and predict daily sales across different stores. The key features of this dataset include:

    Store ID: Each store has a unique identifier which helps in distinguishing between the various stores in the dataset.
    Sales: This is the main target variable representing the daily sales amount for each store. The goal is to predict this value based on other features.
    Customers: The number of customers who visited the store on a particular day. This can provide insights into customer behavior and store traffic.
    Date: The specific date on which the sales and other observations were recorded. This feature is crucial for time series analysis.
    DayOfWeek: The day of the week (ranging from 1 to 7) on which the sales were recorded. This helps in analyzing weekly patterns in sales.
    Open: A binary indicator showing whether the store was open (1) or closed (0) on that day.
    Promo: Indicates whether the store was running a promotional campaign on that day (1 for yes, 0 for no).
    StateHoliday: Indicates whether the day was a state holiday. Different codes (like 'a', 'b', 'c') represent different types of holidays.
    SchoolHoliday: A binary indicator showing whether the day was a school holiday (1 for yes, 0 for no). School holidays can affect sales due to changes in shopping behavior.

Store Dataset

The store dataset provides additional information about each store, which complements the daily sales data. The key features of this dataset include:

    Store ID: A unique identifier for each store, which can be used to link this dataset with the Rossmann Stores Data.
    StoreType: Categorical variable representing the type of store (e.g., 'a', 'b', 'c', 'd'). Different types may indicate varying sizes or formats of stores.
    Assortment: Describes the level of assortment available at the store (e.g., 'a' for basic, 'b' for extra, 'c' for extended). This can influence customer choice and sales.
    CompetitionDistance: The distance to the nearest competitor store, measured in meters. Closer competition might impact the stores sales.
    CompetitionOpenSinceMonth: The month when the nearest competitor store opened. This provides context about the competitive environment.
    CompetitionOpenSinceYear: The year when the nearest competitor store opened.
    Promo2: A binary indicator showing whether the store participates in Promo2, which is a continuous and recurring promotional activity (1 for yes, 0 for no).
    Promo2SinceWeek: The calendar week when the store started participating in Promo2. This helps in understanding the duration and impact of the promotion.
    Promo2SinceYear: The year when the store started participating in Promo2.
    PromoInterval: Describes the months during which Promo2 is active (e.g., 'Jan,Apr,Jul,Oct' means Promo2 is active in January, April, July, and October).

### Hypothetical Statement - 1

Research Hypothesis (Alternate Hypothesis): There is a significant difference in sales between stores that are open and closed on Sundays.

Null Hypothesis: There is no significant difference in sales between stores that are open and closed on Sundays.

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer : To test this hypothesis, we can use a two-sample t-test to compare the mean sales of stores that are open and closed on Sundays.

#### 2. Perform an appropriate statistical test.

Let's conduct a two-sample t-test to compare the mean sales of open and closed stores on Sundays.

````python
import scipy.stats as stats

# Separate data for stores open and closed on Sundays
open_sunday_sales = total_df[total_df['DayOfWeek'] == 7]['Sales']  # Sales for stores open on Sundays
closed_sunday_sales = total_df[total_df['DayOfWeek'] != 7]['Sales']  # Sales for stores closed on Sundays

# Perform two-sample t-test
t_statistic, p_value = stats.ttest_ind(open_sunday_sales, closed_sunday_sales)  # Perform t-test

# Print the results
print("T-statistic:", t_statistic)  # Print the t-statistic
print("P-value:", p_value)  # Print the p-value
````

##### Which statistical test have you done to obtain P-Value?

Answer : two-sample t-test

##### Why did you choose the specific statistical test?

Answer : We are comparing the means of two independent groups (stores open and closed on Sundays).
    The t-test assumes that the samples are normally distributed, which is reasonable given the large sample size.
    It is appropriate for comparing the means of continuous variables (sales) between two groups.

If the p-value is less than a predetermined significance level (e.g., 0.05), we reject the null hypothesis and conclude that there is a significant difference in sales between stores that are open and closed on Sundays.

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize outliers using a boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x=total_df['CompetitionDistance'])
plt.title('Boxplot for CompetitionDistance')
plt.show()

# Cap outliers in 'CompetitionDistance'
q1 = total_df['CompetitionDistance'].quantile(0.25)  # Calculate the first quartile
q3 = total_df['CompetitionDistance'].quantile(0.75)  # Calculate the third quartile
iqr = q3 - q1  # Calculate the interquartile range (IQR)
lower_bound = q1 - 1.5 * iqr  # Calculate the lower bound for outliers
upper_bound = q3 + 1.5 * iqr  # Calculate the upper bound for outliers

# Cap outliers by replacing them with the upper or lower bound
total_df['CompetitionDistance'] = np.where(total_df['CompetitionDistance'] > upper_bound, upper_bound,
                                           np.where(total_df['CompetitionDistance'] < lower_bound, lower_bound, total_df['CompetitionDistance']))
````

        Steps:
            Calculate the first quartile (Q1) and the third quartile (Q3).
            Compute the IQR as the difference between Q3 and Q1.
            Determine the lower bound as Q11.5IQRQ11.5IQR and the upper bound as Q3+1.5IQRQ3+1.5IQR.

        Reason: The IQR method is a robust way to identify and handle outliers because it is not affected by extreme values and focuses on the central 50% of the data. Capping the outliers instead of removing them preserves the data structure and ensures that the analysis remains comprehensive.

Why These Techniques Were Used:

    Boxplot Visualization: This provides a clear and immediate way to see where outliers exist and how they compare to the rest of the data. It is a standard and widely understood method in exploratory data analysis.
    IQR Method: This is a commonly used technique to handle outliers as it is straightforward and based on the statistical distribution of the data. #### 1. Expand Contraction

````python
# Define a dictionary of contractions
contractions = {
    "ain't": "am not",
    "aren't": "are not",
    "can't": "cannot",
    "can't've": "cannot have",
    "'cause": "because",
    "could've": "could have",
    "couldn't": "could not",
    "couldn't've": "could not have",
    "didn't": "did not",
}

import re

def expand_contractions(text, contraction_mapping=contractions):
    """
    Expand contractions in text using a dictionary of contractions.
    """
    # Compile a regular expression pattern to match contractions
    contractions_pattern = re.compile('({})'.format('|'.join(contraction_mapping.keys())),
                                      flags=re.IGNORECASE|re.DOTALL)

    # Define a function to expand each matched contraction
    def expand_match(contraction):
        match = contraction.group(0)
        first_char = match[0]
        # Retrieve the expanded form of the contraction from the dictionary
        expanded = contraction_mapping.get(match) if contraction_mapping.get(match) else contraction_mapping.get(match.lower())
        expanded = first_char + expanded[1:]  # Preserve the case of the original contraction
        return expanded

    # Use the sub() method to replace contractions with their expanded forms
    expanded_text = contractions_pattern.sub(expand_match, text)
    return expanded_text

# Test the function with a sample text
text = "I can't believe it's Friday already."
expanded_text = expand_contractions(text)
print(expanded_text)
````

````python
import re
import pandas as pd

def remove_urls(text):
    """
    Remove URLs from a string.

    Parameters:
    - text (str): Input text from which to remove URLs.

    Returns:
    - str: Text with URLs removed.
    """
    # Use regular expression to remove URLs starting with http, https, or www
    return re.sub(r'http\S+|www\S+|https\S+', '', text)

def remove_words_with_digits(text):
    """
    Remove words containing digits from a string.

    Parameters:
    - text (str): Input text from which to remove words containing digits.

    Returns:
    - str: Text with words containing digits removed.
    """
    # Use regular expression to remove words containing digits
    return re.sub(r'\b\w*\d\w*\b', '', text)

# Example data containing sentences with URLs and digits
data = {
    'text': [
        'This is the test sentence with a URL: https://www.example.com',
        'Another sentence with numbers 123!',
        'Text without any URLs or digits.',
        'http://urlwithoutwww.com'
    ]
}

# Create a DataFrame from the example data
total_df = pd.DataFrame(data)

# Apply the remove_urls function to the 'text' column to remove URLs
total_df['text'] = total_df['text'].apply(remove_urls)

# Apply the remove_words_with_digits function to the 'text' column to remove words with digits
total_df['text'] = total_df['text'].apply(remove_words_with_digits)

# Print the resulting DataFrame to see the changes
print(total_df)
````

````python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import pandas as pd

# Download the stopwords and punkt packages from NLTK
nltk.download('stopwords')
nltk.download('punkt')

# Create a set of English stopwords
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    """
    Remove stopwords from a given text.

    Parameters:
    - text (str): Input text from which to remove stopwords.

    Returns:
    - str: Text with stopwords removed.
    """
    # Tokenize the input text into words
    words = word_tokenize(text)

    # Filter out the stopwords from the tokenized words
    filtered_words = [word for word in words if word.lower() not in stop_words]

    # Join the filtered words back into a single string
    return ' '.join(filtered_words)

# Example data containing sentences with stopwords
data = {
    'text': [
        'This is a sample text with some common English stopwords.',
        'Stopwords are common words that are often removed during text analysis.',
        'We need to remove them to focus on more meaningful words.',
    ]
}

# Create a DataFrame from the example data
total_df = pd.DataFrame(data)

# Apply the remove_stopwords function to the 'text' column to remove stopwords
total_df['text'] = total_df['text'].apply(remove_stopwords)

# Print the resulting DataFrame to see the changes
print(total_df)
````
